### Filling Categorical Missing Values

From the data description, we can observe that some of the 'NA' are an option and not missing values. 
So for those we'll change 'NA' to 'None' and for the rest we'll use Mode

```python
# Impute using a constant value
for column in [
    'Alley',
    'BsmtQual',
    'BsmtCond',
    'BsmtExposure',
    'BsmtFinType1',
    'BsmtFinType2',
    'FireplaceQu',
    'GarageType',
    'GarageFinish',
    'GarageQual',
    'GarageCond',
    'PoolQC',
    'Fence',
    'MiscFeature'
]:
    data2[column] = data2[column].fillna("None")

# Impute using the column mode
for column in [
    'MSZoning',
    'Utilities',
    'Exterior1st',
    'Exterior2nd',
    'MasVnrType',
    'Electrical',
    'KitchenQual',
    'Functional',
    'SaleType'
]:
    data2[column] = data2[column].fillna(data2[column].mode()[0])
```

### Filling Numeric missing values

```python
data3.select_dtypes(np.number).loc[:, data2.isna().sum() > 0].columns
```

For numeric values mean imputation is one way to deal with missing values.
But we'll do KNN imputation.

KNN imputation looks around the missing value to predict what the value should be.

```python
def knn_impute(df, na_target):
    df = df.copy()
    
    #just the numeric columns to be used in knn
    numeric_df = df.select_dtypes(np.number)
    
    #cols with no missing values in numeric df
    non_na_columns = numeric_df.loc[: ,numeric_df.isna().sum() == 0].columns
    
    #all values of the target(missing value col) that are not missing
    y_train = numeric_df.loc[numeric_df[na_target].isna() == False, na_target]
    
    #all rest of the data with no missing values
    X_train = numeric_df.loc[numeric_df[na_target].isna() == False, non_na_columns]
    
    #all values of target with missing values
    X_test = numeric_df.loc[numeric_df[na_target].isna() == True, non_na_columns]
    
    knn = KNeighborsRegressor()
    knn.fit(X_train, y_train)
    
    y_pred = knn.predict(X_test)
    
    #put the predicted values in missing values
    df.loc[df[na_target].isna() == True, na_target] = y_pred
    
    return df
```

```python
for column in [
    'LotFrontage',
    'MasVnrArea',
    'BsmtFinSF1',
    'BsmtFinSF2',
    'BsmtUnfSF',
    'TotalBsmtSF',
    'BsmtFullBath',
    'BsmtHalfBath',
    'GarageYrBlt',
    'GarageCars',
    'GarageArea'
]:
    data3 = knn_impute(data3, column)
```

## Feature Engineering

We will combine some features that can be combined for better results.

```python
data4["SqFtPerRoom"] = data4["GrLivArea"] / (data4["TotRmsAbvGrd"] +
                                                       data4["FullBath"] +
                                                       data4["HalfBath"] +
                                                       data4["KitchenAbvGr"])

data4['Total_Home_Quality'] = data4['OverallQual'] + data4['OverallCond']

data4['Total_Bathrooms'] = (data4['FullBath'] + (0.5 * data4['HalfBath']) +
                               data4['BsmtFullBath'] + (0.5 * data4['BsmtHalfBath']))

data4["HighQualSF"] = data4["1stFlrSF"] + data4["2ndFlrSF"]
```

## Feature Transformation

Feature Transformation is needed when the data is skewed. Lets check which all columns are skewed and need transformation.

```python
import scipy.stats
```

```python
#skewness only works for numeric column

skew_df = pd.DataFrame(data5.select_dtypes(np.number).columns, columns=['Feature'])
skew_df['Skew'] = skew_df['Feature'].apply(lambda feature: scipy.stats.skew(data5[feature]))
skew_df['Absolute Skew'] = skew_df['Skew'].apply(abs)
skew_df['Skewed'] = skew_df['Absolute Skew'].apply(lambda x: True if x >= 0.5 else False)
skew_df
```

```python
skew_df.query("Skewed == True")['Feature'].values
```

Using Log transformation to correct skewness. But We'll use log(x+1) transformation to deal with 0 values.

```python
for column in skew_df.query("Skewed == True")['Feature'].values:
    data5[column] = np.log1p(data5[column])
```

We can also observe that we have cyclic data 'MoSold'. 
And to deal with that we'll use -cos with a constant value that puts 1 and 12 close to each other as month should be.

```python
data4['MoSold'] = (-np.cos(0.5236 * data5['MoSold']))
```

## Bagging Ensemble

We will use bagging ensemble to get better results. We will use top 5 models we got from  pycaret for ensembling.

```python
#Importing all models

from catboost import CatBoostRegressor
from sklearn.linear_model import BayesianRidge, HuberRegressor, Ridge, OrthogonalMatchingPursuit
from lightgbm import LGBMRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python

models = {
    "catboost": CatBoostRegressor(verbose=0),
    "br": BayesianRidge(),
    "lightgbm": LGBMRegressor(),
    "ridge": Ridge(),
    "omp": OrthogonalMatchingPursuit()
}
```

```python
for name, model in models.items():
    model.fit(train_final, log_target)
    print(name + " trained.")
```

## Evaluate

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
results = {}
scores = {}
kf = KFold(n_splits=10)

for name, model in models.items():
    
    #using exp to convert back from log transformation
    result = np.exp(np.sqrt(-cross_val_score(model, train_final, log_target, scoring='neg_mean_squared_error', cv=kf)))
    
    score = cross_val_score(model, train_final, log_target, cv=kf, scoring='accuracy')
    
    results[name] = result
    scores[name] = score
```

## Combine Predictions

```python
# We will give weightage according to accuracy of model
#started with giving each 0.2
#later changed it as this was giving better result
final_predictions = (
    0.4 * np.exp(models['catboost'].predict(test_final)) +
    0.2 * np.exp(models['br'].predict(test_final)) +
    0.2 * np.exp(models['lightgbm'].predict(test_final)) +
    0.1 * np.exp(models['ridge'].predict(test_final)) +
    0.1 * np.exp(models['omp'].predict(test_final))
)
```
